## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

- **Pertanyaan 1:** Apa kategori produk yang paling banyak terjual (berdasarkan jumlah pesanan) selama tahun 2017?
- **Pertanyaan 2:** Apa jenis metode pembayaran yang paling sering digunakan untuk transaksi di tahun 2017?

**Q1 [SMART Method]**
*   S (Specific): Fokus pada volume penjualan per kategori produk.
*   M (Measurable): Bisa dihitung dari jumlah baris atau ID pesanan.
*   A (Action-oriented): Hasilnya membantu menentukan fokus stok barang.
*   R (Relevant): Penting untuk mengetahui minat pasar.
*   T (Time-bound): Dibatasi pada tahun 2017.

**Q2 [SMART Method]**
*   S (Specific): Fokus pada preferensi cara bayar pelanggan.
*   M (Measurable): Bisa dihitung dari frekuensi setiap jenis pembayaran.
*   A (Action-oriented): Membantu tim keuangan/IT mengoptimalkan gerbang pembayaran tersebut.
*   R (Relevant): Penting untuk kemudahan transaksi pelanggan.
*   T (Time-bound): Dibatasi pada 2017.

### Cleaning Data

#### Fixing ... problem

```python
# Melakukan konversi kolom waktu dan mengubahnya menjadi tipe datetime
datetime_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"]
for column in datetime_columns:
    df_orders[column] = pd.to_datetime(df_orders[column])

# Mengisi kategori yang kosong dengan 'unknown' pada tabel products
df_products["product_category_name"] = df_products["product_category_name"].fillna("unknown")

# Menggabungkan orders, order_items, produk, dan translasi nama kategori untuk menjawab pertanyaan no 1
df_q1 = pd.merge(left=df_order_items, right=df_orders, how="left", left_on="order_id", right_on="order_id")
df_q1 = pd.merge(left=df_q1, right=df_products, how="left", left_on="product_id", right_on="product_id")
df_q1 = pd.merge(left=df_q1, right=df_product_category_name_translation, how="left", left_on="product_category_name", right_on="product_category_name")

# Mengganungkan tabel orders dan payments untuk menjawab soal no 2
df_q2 = pd.merge(left=df_order_payments, right=df_orders, how="left", left_on="order_id", right_on="order_id")

# Melakukan filter untuk menampilkan data tahun 2017 dengan status 'delivered'
# Filter Q1 (Kategori)
main_df_q1 = df_q1[(df_q1["order_purchase_timestamp"].dt.year == 2017) & (df_q1["order_status"] == "delivered")]
# Filter Q2 (Pembayaran)
main_df_q2 = df_q2[(df_q2["order_purchase_timestamp"].dt.year == 2017) & (df_q2["order_status"] == "delivered")]

print("Cleaning Selesai!")
print(f"Jumlah baris data kategori 2017: {main_df_q1.shape[0]}")
print(f"Jumlah baris data pembayaran 2017: {main_df_q2.shape[0]}")
```

**Insight:** (Opsional)
- Kesalahan tipe data pada kolom order_purchase_timestamp, order_approved_at, dan delivery date berhasil diubah menjadi tipe datetime.
- Missing value yang ada pada Kolom product_category_name diubah menjadi 'unknown' agar tidak menggangu proses analisis.
- Dilakukan penggabungan tabel orders, order_item, dan product untuk menjawab pertanyaan no 1 (df_q1).
- Dilakukan penggabungan tabel orders, dan payment untuk menjawab pertanyaan no 2 (df_q2).
- Dilakukan filtering untuk pesanan yang memiliki status 'delivered' agar analisis dapat berfokus pada pesanan yang sudah diterima.

## Analisis Lanjutan (Opsional)

```python
# Menyiapkan data RFM
rfm_data = pd.merge(left=main_df_q1, right=df_customers, how="left", on="customer_id")

# Menentukan tanggal referensi (1 hari setelah transaksi terakhir)
snapshot_date = rfm_data['order_purchase_timestamp'].max() + pd.Timedelta(days=1)

# Membuat dataframe RFM
rfm_df = rfm_data.groupby(by="customer_unique_id", as_index=False).agg({
    "order_purchase_timestamp": lambda x: (snapshot_date - x.max()).days, # Recency
    "order_id": "nunique", # Frequency
    "price": "sum" # Monetary
})

# Rename kolom
rfm_df.columns = ["customer_id", "recency", "frequency", "monetary"]

# Visualisasi RFM
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 10))
colors = ["#72BCD4"] * 5

# Berdasarkan Recency
sns.barplot(y="recency", x="customer_id", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, ax=ax[0])
ax[0].set_title("By Recency (days)", fontsize=18)
ax[0].tick_params(axis='x', rotation=45)

# Berdasarkan Frequency
sns.barplot(y="frequency", x="customer_id", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, ax=ax[1])
ax[1].set_title("By Frequency", fontsize=18)
ax[1].tick_params(axis='x', rotation=45)

# Berdasarkan Monetary
sns.barplot(y="monetary", x="customer_id", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, ax=ax[2])
ax[2].set_title("By Monetary", fontsize=18)
ax[2].tick_params(axis='x', rotation=45)

plt.suptitle("Best Customer Based on RFM Parameters", fontsize=25)
plt.show()
```
